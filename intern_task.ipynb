*Посмотрим на коэффициенты корреляции между фичами*

*Уберем те столбцы, которые много с кем коррелируют*

```python
df_corr_1 = df_corr > 0.7
df_uncorr = df_corr_1.loc[:, df_corr_1.sum() < 15]
column_set = set(df_uncorr.columns)
df_unc = df[column_set]
df_unc
```

```python
df_unc['rank'] = df['rank']
df_unc
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, random_state = 52, train_size = 0.8)
```

```python
x_train = train.drop('rank', axis = 1)
y_train = train['rank']

x_test = test.drop('rank', axis = 1)
y_test = test['rank']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import ndcg_score
from sklearn.metrics import label_ranking_average_precision_score
from sklearn.metrics import dcg_score
```

```python
y_pred = np.asarray(y_pred)
y_test = np.asarray(y_test)
```

```python
y_pred_reshape = y_pred.reshape(1, y_pred.shape[0])
y_test_reshape = y_test.reshape(1, y_test.shape[0])
```

```python
ndcg_score(y_test_reshape, y_pred_reshape)
```
